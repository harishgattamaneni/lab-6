```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
import numpy as np

# Read the data from the Excel file (assuming you've already read the data)
df = pd.read_excel('/Users/harishgattamaneni/Downloads/embeddingsdata.xlsx', engine='openpyxl')

# Extract the two features
independent_variable = df['embed_0']  # Use Feature1 as the independent variable
dependent_variable = df['embed_1']    # Use Feature2 as the dependent variable

# Reshape the data for sklearn (needed when working with a single feature)
independent_variable = np.array(independent_variable).reshape(-1, 1)

# Create and train the linear regression model
model = LinearRegression()
model.fit(independent_variable, dependent_variable)

# Predict the values using the model
predicted_values = model.predict(independent_variable)

# Calculate the mean squared error
mse = mean_squared_error(dependent_variable, predicted_values)

# Round the MSE to an integer
rounded_mse = round(mse)

print(f"Mean Squared Error: {mse}")
print(f"Rounded Mean Squared Error (to the nearest integer): {rounded_mse}")
```

```python
coefficients = model.coef_
intercept = model.intercept_

# Make predictions
y_pred = model.predict(independent_variable)

# Calculate mean squared error
mse = mean_squared_error(dependent_variable, y_pred)

# Calculate the coefficient of determination (R-squared)
r2 = r2_score(dependent_variable, y_pred)

# Print the results
print("Coefficients:", coefficients)
print("Intercept:", intercept)
print("Mean Squared Error:", mse)
print("R-squared:", r2)
```
